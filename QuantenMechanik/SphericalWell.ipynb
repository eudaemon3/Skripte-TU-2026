```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use('science')
plt.rcParams.update({'font.size': 14})

cot = lambda x : 1 / np.tan(x)
```

```python
a = 1
m = 1
hbar = 1
y = np.linspace(0.01, 2*np.pi,1000)

V0_min = (hbar*np.pi / 2 / a)**2 / (2*m)
Ns = [0.5, 1, 2]

fig, ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(y, np.abs(np.where(cot(y) < 0, np.sin(y), 0)), label=r'$|\sin(y)|$', lw=1.5)
for n in Ns: 
    gamma = hbar / (a * np.sqrt(2*m*n*V0_min))
    label_str = f'{n}' + r'$\cdot V_{0,min}$'
    ax.plot(y, gamma*y, linestyle='-.', label=label_str)
ax.grid(True, 'major')
ax.grid(True, 'minor', linestyle='--', alpha=0.5)
ax.legend(frameon=True)
ax.set_xlabel(r'$y = k_1 a$', fontsize=16)
ax.set_ylabel(r'$f(y)$', fontsize=16)
ax.set_title('Kugelsymmetrischer Potenzialtopf', fontsize=18, pad=10)
plt.tight_layout()
plt.show()
```
