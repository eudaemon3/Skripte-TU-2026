```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt
```

## Exercise 1

```python
N_a = 20
k_a = 5

N_b = 50
k_b = 15

p_2 = 0.2 # probability of getting a 6 with biased die
p_2 = 1/6 # probability of getting a 6 with fair die

P_H1 = 1/3 # picking a biased die -> 1 out of 3
P_H2 = 2/3 # picking a fair die -> 2 out of 3

odds_ratio = lambda N, k : p_2**k * (1 - p_2)**(N - k) * P_H1 / (p_2**k * (1 - p_2)**(N - k) * P_H2)

print(f'a) Odds Ratio o_12 = {odds_ratio(N_a, k_a):.3f}')
print(f'b) Odds Ratio o_12 = {odds_ratio(N_b, k_b):.3f}')
```

## Exercise 2

```python
P_S1_given_K1 = 0.97
P_S1_given_K2 = 0.67
P_S2_given_K1 = 0.85
P_S2_given_K2 = 0.52

P_K1 = 1.0e-6
P_K2 = 1.0e-3

BF = P_S1_given_K1 * P_S2_given_K1 / (P_S1_given_K2 * P_S2_given_K2)

o_prior = P_K1 / P_K2
o_post = BF * o_prior

print(f'Bayes factor = {BF:.3f}')
print(f'prior odds ratio = {o_prior:.3f}')
print(f'posterior odds ratio = {o_post:.5f}')
```

## Exercise 3

```python
# Data Experiment 1
N_1 = 100
k_1 = 20

# Data Experiment 2
N_2 = 20 
k_2 = 13

p_2 = 0.8  # claimed accuracy on second experiment
p_1 = 0.65 # claimed accuracy on first experiment
p_0 = 0.5  # assumed accuracy no effect

o_a_21 = p_2**k_2 * (1 - p_2)**(N_2 - k_2) / (p_0**k_2 * (1 - p_0)**(N_2 - k_2))

o_b_21 = o_a_21 * p_1**k_1 * (1 - p_1)**(N_1 - k_1) / (p_0**k_1 * (1 - p_0)**(N_1 - k_1))

print(f'a) Odds Ratio oa_21 = {o_a_21:.3f}')
print(f'b) Odds Ratio ob_21 = {o_b_21:.2e}')
```

```python
N = 10
k = 9

p = 0.9
p_0 = 0.5

P_D_H2 = comb(N, k) * p**k * (1 - p)**(N - k)
P_D_H1 = comb(N, k) * p_0**k * (1 - p_0)**(N - k)

o_21 = P_D_H2 / P_D_H1
print(f'a) Probability p(D|H2) = {P_D_H2:.3f}')
print(f'b) Odds Ratio oa_21 = {o_21:.3f}')
```
