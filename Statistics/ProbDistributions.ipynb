# Exercise 1.1

```python
def hypergeometric(N, K,N_a, K_a):
    return comb(N_a, K_a) * comb(N-N_a, K-K_a) / comb(N,K)

def binomial(k, n, p):
    return comb(n, k) * p**k * (1-p)**(n-k)
```

```python
x_bar = 0.15 + 1 + 3*0.35
y_bar = 0.45 + 0.6 + 0.25*3

xx_bar = 0.15 + 4*0.5 + 9*0.35
yy_bar = 0.45 + 4*0.3 + 9*0.25

var_x = xx_bar - x_bar**2
var_y = yy_bar - y_bar**2
print(f'Var(x) = {var_x:.2f}   Var(y) = {var_y:.2f}')

xy_bar = 1*0.1 + 2*0.3 + 3*0.05 + 4*0.1 + 6*0.2 + 3*0.05 + 6*0.1 + 9*0.1
c = xy_bar - x_bar*y_bar
R = c / np.sqrt(var_x*var_y)

print(f'Covariance C = {c:.2f}   Correlation R = {R:.2f}')
```

# Exercise 5.1

```python
def marginal_prob(P_y_given_x: np.ndarray, P_x_given_y: np.ndarray):

    if P_x_given_y.shape != (2,2) or P_y_given_x.shape != (2,2):
        print("Enter 2 matrices with dimension (2,2)!")
        raise TypeError
    
    v1 = np.allclose(np.sum(P_y_given_x, axis=1), 1)
    v2 = np.allclose(np.sum(P_x_given_y, axis=0), 1)
    if not v1 or not v2:
        print("Enter 2 valid probability matrices!")
        raise ValueError

    A = np.array([
        [ P_y_given_x[0,0], 0, -P_x_given_y[0,0], 0 ],
        [ P_y_given_x[0,1], 0, 0, -P_x_given_y[0,1] ],
        [ 1, 1, 0, 0 ],
        [ 0, 0, 1, 1 ]    
    ])

    b = np.array([0, 0, 1, 1])

    prob = np.linalg.solve(A, b)
    return prob
```

```python
# colums must sum to 1
P_x_given_y = np.array([
    [2/3, 4/7],
    [1/3, 3/7]
])

# rows must sum to 1
P_y_given_x = np.array([
    [2/6, 4/6],
    [1/4, 3/4]
])

p = marginal_prob(P_y_given_x, P_x_given_y)
```

```python
px1, px2, py1, py2 = p
P_xy = np.array([
    [P_y_given_x[0,0]*px1, P_y_given_x[0,1]*px1],
    [P_y_given_x[1,0]*px2, P_y_given_x[1,1]*px2],
])

P_yx = np.array([
    [P_x_given_y[0,0]*py1, P_x_given_y[0,1]*py2],
    [P_x_given_y[1,0]*py1, P_x_given_y[1,1]*py2],
])

print(P_xy)
print('---------------')
print(P_yx)
print('---------------')
print(f'Is valid: {np.allclose(P_xy, P_yx)}')
```
