```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.integrate import quad
```

```python
def func(x, x0, lam):
    x = np.asarray(x)
    return np.where(
        x < x0,
        0,
        np.where(x > x0 + lam, 1, (x - x0) / lam)
    )

x = np.linspace(-4,4, 1000)
plt.plot(x, func(x, 1, 1))
plt.plot(x, func(x, 1, 2))
plt.plot(x, func(x, 0, 2))
plt.grid(True)
plt.show()

```

```python
x0 = 1
lam = 1
x = np.linspace(-5, 5, 1000)

H = np.heaviside
y = lambda x, x0, lam : (H(x-x0, 0.5) - H(x-x0-lam, 0.5))*(x/lam - x0/lam) + H(x-x0-lam, 0.5)

plt.plot(x, y(x, 1, 1))
plt.plot(x, y(x, 1, 2))
plt.plot(x, y(x, 0, 2))
plt.show()
```
